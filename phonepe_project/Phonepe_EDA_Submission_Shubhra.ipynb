# **Project Summary -**

This project focuses on analyzing and visualizing digital payment data from the PhonePe Pulse dataset to extract meaningful business insights. With the rapid growth of digital transactions in India, understanding transaction behavior, user engagement, and insurance adoption is critical for enhancing customer experience and optimizing financial services.

The project begins with extracting structured data from a publicly available GitHub repository containing JSON files organized into three major categories: Aggregated, Map, and Top. These datasets provide information on transactions, users, and insurance activities across different states, districts, and time periods. Using Python, these JSON files were parsed and transformed into structured DataFrames for further analysis.

The extracted datasets were then stored in a SQL database (SQLite), enabling efficient querying and analysis. SQL queries were used to compute key metrics such as total transaction volume by type, yearly growth trends, and top-performing states.

The project follows a structured EDA approach using the UBM rule (Univariate, Bivariate, Multivariate Analysis) to explore the data:

Univariate Analysis examined individual variables such as transaction types and device usage.
Bivariate Analysis explored relationships such as transaction count vs amount and user engagement vs app usage.
Multivariate Analysis combined multiple variables to identify patterns across states and time.

A total of 20+ visualizations were created using Matplotlib and Seaborn to present insights such as:

Growth of digital transactions over time
Dominance of peer-to-peer payments
State-wise transaction distribution
User engagement trends
Device brand preferences
Insurance adoption patterns

The results show a significant increase in digital transactions year over year, with peer-to-peer payments contributing the largest share. States like Maharashtra, Karnataka, and Uttar Pradesh emerged as top contributors. Additionally, a strong correlation was observed between registered users and app opens, indicating high engagement levels.

The project concludes with actionable business insights that can help PhonePe improve marketing strategies, optimize product offerings, and enhance customer retention. A Streamlit dashboard can further be developed to provide real-time interactive insights.

###D. Extract json into DataFrames

####i) Aggregated Transaction

````python
path = "/content/pulse/data/aggregated/transaction/country/india/"

agg_trans_list = []

for year in os.listdir(path):
    year_path = os.path.join(path, year)
    if os.path.isdir(year_path):
        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = file.replace(".json", "")
                file_path = os.path.join(year_path, file)

                with open(file_path, "r") as f:
                    data = json.load(f)

                for item in data["data"]["transactionData"]:
                    name = item["name"]
                    count = item["paymentInstruments"][0]["count"]
                    amount = item["paymentInstruments"][0]["amount"]

                    agg_trans_list.append([year, quarter, name, count, amount])

agg_trans = pd.DataFrame(
    agg_trans_list,
    columns=["Year", "Quarter", "Transaction_type", "Transaction_count", "Transaction_amount"]
)

agg_trans.head()
````

####ii) Aggregated User

````python
path = "/content/pulse/data/aggregated/user/country/india/"

agg_user_list = []

for year in os.listdir(path):
    year_path = os.path.join(path, year)
    if os.path.isdir(year_path):
        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = file.replace(".json", "")
                file_path = os.path.join(year_path, file)

                with open(file_path, "r") as f:
                    data = json.load(f)

                try:
                    for item in data["data"]["usersByDevice"]:
                        brand = item["brand"]
                        count = item["count"]
                        percentage = item["percentage"]

                        agg_user_list.append([year, quarter, brand, count, percentage])
                except:
                    pass

agg_user = pd.DataFrame(
    agg_user_list,
    columns=["Year", "Quarter", "Brand", "User_count", "Percentage"]
)

agg_user.head()
````

####iii) Aggregated Insurance

````python
path = "/content/pulse/data/aggregated/insurance/country/india/"

agg_ins_list = []

for year in os.listdir(path):
    year_path = os.path.join(path, year)
    if os.path.isdir(year_path):
        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = file.replace(".json", "")
                file_path = os.path.join(year_path, file)

                with open(file_path, "r") as f:
                    data = json.load(f)

                if data["data"]["transactionData"] is not None:
                    for item in data["data"]["transactionData"]:
                        name = item["name"]
                        count = item["paymentInstruments"][0]["count"]
                        amount = item["paymentInstruments"][0]["amount"]

                        agg_ins_list.append([year, quarter, name, count, amount])

agg_ins = pd.DataFrame(
    agg_ins_list,
    columns=["Year", "Quarter", "Insurance_type", "Insurance_count", "Insurance_amount"]
)

agg_ins.head()
````

####iv) Map Transaction

````python
path = "/content/pulse/data/map/transaction/hover/country/india/"

map_trans_list = []

for year in os.listdir(path):
    year_path = os.path.join(path, year)
    if os.path.isdir(year_path):
        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = file.replace(".json", "")
                file_path = os.path.join(year_path, file)

                with open(file_path, "r") as f:
                    data = json.load(f)

                for item in data["data"]["hoverDataList"]:
                    state = item["name"]
                    count = item["metric"][0]["count"]
                    amount = item["metric"][0]["amount"]

                    map_trans_list.append([year, quarter, state, count, amount])

map_trans = pd.DataFrame(
    map_trans_list,
    columns=["Year", "Quarter", "State", "Transaction_count", "Transaction_amount"]
)

map_trans.head()
````

####v) Map User

````python
path = "/content/pulse/data/map/user/hover/country/india/"

map_user_list = []

for year in os.listdir(path):
    year_path = os.path.join(path, year)
    if os.path.isdir(year_path):
        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = file.replace(".json", "")
                file_path = os.path.join(year_path, file)

                with open(file_path, "r") as f:
                    data = json.load(f)

                for state, details in data["data"]["hoverData"].items():
                    registered_users = details["registeredUsers"]
                    app_opens = details["appOpens"]

                    map_user_list.append([year, quarter, state, registered_users, app_opens])

map_user = pd.DataFrame(
    map_user_list,
    columns=["Year", "Quarter", "State", "Registered_Users", "App_Opens"]
)

map_user.head()
````

####vi) Map Insurance

````python
path = "/content/pulse/data/map/insurance/hover/country/india/"

map_ins_list = []

for year in os.listdir(path):
    year_path = os.path.join(path, year)
    if os.path.isdir(year_path):
        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = file.replace(".json", "")
                file_path = os.path.join(year_path, file)

                with open(file_path, "r") as f:
                    data = json.load(f)

                if data["data"]["hoverDataList"] is not None:
                    for item in data["data"]["hoverDataList"]:
                        state = item["name"]
                        count = item["metric"][0]["count"]
                        amount = item["metric"][0]["amount"]

                        map_ins_list.append([year, quarter, state, count, amount])

map_ins = pd.DataFrame(
    map_ins_list,
    columns=["Year", "Quarter", "State", "Insurance_count", "Insurance_amount"]
)

map_ins.head()
````

####vii) Top Transaction

````python
path = "/content/pulse/data/top/transaction/country/india/"

top_trans_list = []

for year in os.listdir(path):
    year_path = os.path.join(path, year)
    if os.path.isdir(year_path):
        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = file.replace(".json", "")
                file_path = os.path.join(year_path, file)

                with open(file_path, "r") as f:
                    data = json.load(f)

                for item in data["data"]["states"]:
                    state = item["entityName"]
                    count = item["metric"]["count"]
                    amount = item["metric"]["amount"]

                    top_trans_list.append([year, quarter, state, count, amount])

top_trans = pd.DataFrame(
    top_trans_list,
    columns=["Year", "Quarter", "State", "Transaction_count", "Transaction_amount"]
)

top_trans.head()
````

viii) Top User

````python
path = "/content/pulse/data/top/user/country/india/"

top_user_list = []

for year in os.listdir(path):
    year_path = os.path.join(path, year)
    if os.path.isdir(year_path):
        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = file.replace(".json", "")
                file_path = os.path.join(year_path, file)

                with open(file_path, "r") as f:
                    data = json.load(f)

                for item in data["data"]["states"]:
                    state = item["name"]
                    registered_users = item["registeredUsers"]

                    top_user_list.append([year, quarter, state, registered_users])

top_user = pd.DataFrame(
    top_user_list,
    columns=["Year", "Quarter", "State", "Registered_Users"]
)

top_user.head()
````

####viii) Top Insurance

````python
path = "/content/pulse/data/top/insurance/country/india/"

top_ins_list = []

for year in os.listdir(path):
    year_path = os.path.join(path, year)
    if os.path.isdir(year_path):
        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = file.replace(".json", "")
                file_path = os.path.join(year_path, file)

                with open(file_path, "r") as f:
                    data = json.load(f)

                if data["data"]["states"] is not None:
                    for item in data["data"]["states"]:
                        state = item["entityName"]
                        count = item["metric"]["count"]
                        amount = item["metric"]["amount"]

                        top_ins_list.append([year, quarter, state, count, amount])

top_ins = pd.DataFrame(
    top_ins_list,
    columns=["Year", "Quarter", "State", "Insurance_count", "Insurance_amount"]
)

top_ins.head()
````
