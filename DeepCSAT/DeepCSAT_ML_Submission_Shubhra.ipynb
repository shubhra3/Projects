# **Project Summary -**

Customer Satisfaction (CSAT) is one of the most important performance indicators for e-commerce platforms. High CSAT leads to increased customer loyalty, repeat purchases, and better brand reputation. In contrast, poor service experience can negatively impact customer retention and revenue. Companies therefore need tools that can predict customer satisfaction early and identify operational factors that influence service quality.

This project aims to build a machine learning and deep learning system that predicts Customer Satisfaction (CSAT) scores based on service interaction data from an e-commerce platform. The dataset includes information about customer service interactions such as communication channel, category of issue, response times, product information, agent information, and handling time.

The project follows a complete machine learning pipeline, including data exploration, data cleaning, feature engineering, visualization, statistical testing, model training, and evaluation. 1. Linear Regression

3. Gradient Boosting Regression

4. Deep Learning Neural Network

Hyperparameter tuning and cross-validation techniques are used to improve model performance. Evaluation metrics such as MAE, RMSE, and R² score are used to measure prediction accuracy.

The final model provides actionable insights such as:

a. Response time impact on customer satisfaction

b. Channel effectiveness (chat, email, call)

c. Agent performance influence

d. Product category service issues

These insights help businesses improve customer support operations and proactively manage customer experience.

The final model can be deployed as a Streamlit application where users can input interaction details and receive predicted CSAT scores.

### What all manipulations have you done and insights you found?

Manipulations & Insights

Missing values in the dataset were handled using the forward fill (ffill) technique, where missing entries were replaced with the previous valid value in the column. This approach helps maintain data continuity and ensures that the dataset remains usable for analysis and modeling without dropping a large number of rows.

b. CSAT Score Cleaning

The CSAT Score variable represents customer satisfaction ratings that should fall within the valid range of 1 to 5. To ensure data quality and prevent incorrect values from affecting the analysis, the CSAT scores were cleaned using the clip() function, which restricts values within the specified range.

c. Categorical Variable Preparation

Categorical variables such as channel_name, category, Product_category, Agent_name, Supervisor, Manager, Tenure Bucket, and Agent Shift were prepared for analysis. These variables help identify patterns across different service channels, product types, and support teams, and can later be encoded for machine learning models.

d. Datetime Feature Preparation

The dataset contains several time-related variables such as order_date_time, Issue_reported at, issue_responded, and Survey_response_Date. Statement

This hypothesis was derived from the CSAT by Channel box plot, which suggested that some channels may produce higher satisfaction scores than others.

Null and Alternate Hypothesis

Null Hypothesis (H₀):
The mean CSAT score is the same across all service channels.

Alternate Hypothesis (H₁):
The mean CSAT score differs across at least one service channel.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Select relevant columns and remove missing values
anova_df = df[['channel_name','CSAT Score']].dropna()

# Create groups for each channel
channel_groups = [
    group['CSAT Score'].values
    for name, group in anova_df.groupby('channel_name')
]

# Perform One Way ANOVA
f_stat, p_value = f_oneway(*channel_groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

# Significance level
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis: CSAT differs significantly across channels.")
else:
    print("Fail to reject the Null Hypothesis: No significant difference in CSAT across channels.")
````

One-Way ANOVA

#### What all missing value imputation techniques have you used and why did you use those techniques?

Several missing value handling techniques were used to ensure the dataset remains suitable for analysis and machine learning modeling.

1. Forward Fill (ffill)
Forward fill was used to propagate the last valid observation to fill missing values. This method helps maintain data continuity, especially when missing values appear sequentially in records.

2. Median Imputation for Numerical Variables
Numerical columns such as Item_price and connected_handling_time were filled using the median value. Median imputation is preferred over mean when the data contains outliers or skewed distributions because it is less sensitive to extreme values.

3. Mode Imputation for Categorical Variables
Categorical variables such as Customer_City and Product_category were filled using the mode (most frequent value). This ensures that missing categorical entries are replaced with the most representative category in the dataset.

These techniques allow the dataset to retain the maximum number of records while reducing bias and maintaining statistical consistency for machine learning models.

###2. Datetime Feature Engineering

````python
# -------------------------------------
# 2. Datetime Feature Engineering
# -------------------------------------

# Convert datetime columns
date_cols = ['order_date_time','Issue_reported at','issue_responded']

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Create response_time feature
if 'Issue_reported at' in df.columns and 'issue_responded' in df.columns:
    df['response_time'] = (df['issue_responded'] - df['Issue_reported at']).dt.total_seconds()

# Fill response_time missing values
df['response_time'] = df['response_time'].fillna(df['response_time'].median())
````

### 3. Handling Outliers

````python
# -------------------------------------
# 3. Outlier Treatment using IQR
# -------------------------------------

def cap_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return np.where(column > upper, upper, np.where(column < lower, lower, column))

df['Item_price'] = cap_outliers(df['Item_price'])
df['connected_handling_time'] = cap_outliers(df['connected_handling_time'])
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

To handle extreme values in numerical variables, the Interquartile Range (IQR) method was used.

1. IQR Method for Outlier Detection
The IQR method identifies outliers based on the spread of the data. Values falling below Q1 − 1.5 × IQR or above Q3 + 1.5 × IQR are considered outliers.

2. Outlier Capping (Winsorization)
Instead of removing rows containing outliers, the values were capped within the acceptable range using the calculated upper and lower bounds. This method preserves the dataset size while reducing the influence of extreme values on machine learning models.

Why this approach was chosen

Removing rows could significantly reduce the dataset size.

Capping maintains data integrity.

It prevents models such as Linear Regression and Neural Networks from being heavily influenced by extreme values.

After applying outlier treatment, the numerical features become more stable and suitable for model training.

````python
# Initialize scaler
scaler = StandardScaler()

# Drop non-numeric columns that contain lists (tokens and pos_tags)
# Make a copy to avoid SettingWithCopyWarning
X_train_scaled_df = X_train.drop(columns=['tokens', 'pos_tags'], errors='ignore').copy()
X_test_scaled_df = X_test.drop(columns=['tokens', 'pos_tags'], errors='ignore').copy()

# Scale training and test data
X_train_scaled = scaler.fit_transform(X_train_scaled_df)
X_test_scaled = scaler.transform(X_test_scaled_df)

print(X_train_scaled.shape)
print(X_test_scaled.shape)
````

````python
# Build ANN model
ann_model = Sequential()

# Input layer + hidden layers
ann_model.add(Dense(128, activation='relu', input_dim=X_train_scaled.shape[1]))
ann_model.add(Dropout(0.2))

ann_model.add(Dense(64, activation='relu'))
ann_model.add(Dropout(0.2))

ann_model.add(Dense(32, activation='relu'))

# Output layer
ann_model.add(Dense(1, activation='linear'))

# Compile model
ann_model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

# Fit the model
history = ann_model.fit(
    X_train_scaled,
    y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    verbose=1
)

# Predict
y_pred_ann = ann_model.predict(X_test_scaled).flatten()

# Evaluation
mae_ann = mean_absolute_error(y_test, y_pred_ann)
rmse_ann = np.sqrt(mean_squared_error(y_test, y_pred_ann))
r2_ann = r2_score(y_test, y_pred_ann)

print("ANN MAE:", mae_ann)
print("ANN RMSE:", rmse_ann)
print("ANN R2 Score:", r2_ann)
````

For deep learning models, hyperparameter tuning usually involves adjusting:

Here we use EarlyStopping to improve model generalization.

````python
#ANN Hyperparameter Optimization
# Early stopping
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Rebuild model
ann_tuned = Sequential()

ann_tuned.add(Dense(128, activation='relu', input_dim=X_train_scaled.shape[1]))
ann_tuned.add(Dropout(0.3))

ann_tuned.add(Dense(64, activation='relu'))
ann_tuned.add(Dropout(0.3))

ann_tuned.add(Dense(32, activation='relu'))
ann_tuned.add(Dense(1, activation='linear'))

ann_tuned.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

# Fit the Algorithm
history_tuned = ann_tuned.fit(
    X_train_scaled,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

# Predict on the model
y_pred_ann_tuned = ann_tuned.predict(X_test_scaled).flatten()

# Evaluation
mae_ann_tuned = mean_absolute_error(y_test, y_pred_ann_tuned)
rmse_ann_tuned = np.sqrt(mean_squared_error(y_test, y_pred_ann_tuned))
r2_ann_tuned = r2_score(y_test, y_pred_ann_tuned)

print("Tuned ANN MAE:", mae_ann_tuned)
print("Tuned ANN RMSE:", rmse_ann_tuned)
print("Tuned ANN R2 Score:", r2_ann_tuned)
````
